## 🎨 Step 3 — Visual Styling

```python
from typing import Dict

CATEGORY_COLORS = {
    "EU_AI_ACT": "#4F6BED",
    "UK_GDPR":   "#E05A5A",
    "UK_AI_WP":  "#E8913A",
    "NIST":      "#5AB88A",
    "ARTEFACT":  "#A855F7",
    "SYSTEM":    "#F59E0B",
}

EDGE_COLORS = {
    "REQUIRES":    "#EF4444",
    "TO_INCLUDE":  "#F97316",
    "ALIGNS_WITH": "#3B82F6",
    "IMPLEMENTS":  "#8B5CF6",
    "RELATED_TO":  "#6B7280",
    "SUBJECT_TO":  "#F59E0B",
    "INFORMS":     "#10B981",
}

def node_color_mapping(node: Dict) -> str:
    return CATEGORY_COLORS.get(node["properties"].get("category", ""), "#CBD5E1")

def node_scale_mapping(node: Dict) -> float:
    sizes = {"EU_AI_ACT": 1.2, "UK_GDPR": 1.15, "UK_AI_WP": 1.1,
             "NIST": 1.1, "ARTEFACT": 1.3, "SYSTEM": 1.4}
    return sizes.get(node["properties"].get("category", ""), 1.0)

def edge_color_mapping(edge: Dict) -> str:
    return EDGE_COLORS.get(edge["properties"].get("label", ""), "#94A3B8")

print("Styling defined. Legend:")
for cat, col in CATEGORY_COLORS.items():
    print(f"  {col}  {cat}")
print()
print("Edge relationship types:")
for lbl, col in EDGE_COLORS.items():
    print(f"  {col}  {lbl}")
```

## 🌐 Step 4 — Full Interactive Regulatory Graph

> **Touch & mouse friendly** — drag nodes, pinch to zoom, tap to inspect. Click any node to see its full regulatory description in the sidebar.

| Colour | Framework |
|--------|----------|
| 🔵 Indigo | EU AI Act Articles |
| 🔴 Red | UK GDPR |
| 🟠 Amber | UK AI White Paper |
| 🟢 Teal | NIST AI RMF |
| 🟣 Purple | Automation Artefacts |
| 🟡 Gold | AI Systems |


```python
from yfiles_jupyter_graphs import GraphWidget

def build_graph(nodes, edges):
    """Build and return a configured yfiles GraphWidget."""
    w = GraphWidget()
    w.directed = True

    w.nodes = [
        {
            "id": n["id"],
            "properties": {
                "label":       n["label"],
                "category":    n["category"],
                "law":         n["law"],
                "use_case":    n["use_case"],
                "description": n["description"],
                "url":         n["url"],
            }
        }
        for n in nodes
    ]
    w.edges = [
        {
            "id":    e["id"],
            "start": e["start"],
            "end":   e["end"],
            "properties": {"label": e["label"]}
        }
        for e in edges
    ]

    w.set_node_color_mapping(node_color_mapping)
    w.set_node_scale_factor_mapping(node_scale_mapping)
    w.set_edge_color_mapping(edge_color_mapping)
    return w

w_full = build_graph(NODES, EDGES)
print(f"Displaying full graph: {len(NODES)} nodes, {len(EDGES)} edges")
print("Interact: drag nodes | scroll/pinch to zoom | click node for details")
print()
display(w_full)
```

## 🔍 Step 5 — Subgraph Explorer

Filter the graph by regulatory framework or AI system to focus your compliance analysis.

```python
def subgraph_by_category(category_list, title="Subgraph"):
    node_ids = {n["id"] for n in NODES if n["category"] in category_list}
    filtered_nodes = [n for n in NODES if n["id"] in node_ids]
    filtered_edges = [e for e in EDGES
                      if e["start"] in node_ids and e["end"] in node_ids]
    print(f"  {title}: {len(filtered_nodes)} nodes, {len(filtered_edges)} edges")
    display(build_graph(filtered_nodes, filtered_edges))

def subgraph_for_system(system_id, title="System Subgraph"):
    connected = {system_id}
    for e in EDGES:
        if e["start"] == system_id:
            connected.add(e["end"])
        elif e["end"] == system_id:
            connected.add(e["start"])
    # Pull second-degree artefact -> law links
    for e in EDGES:
        if e["start"] in connected:
            connected.add(e["end"])
    filtered_nodes = [n for n in NODES if n["id"] in connected]
    filtered_edges = [e for e in EDGES
                      if e["start"] in connected and e["end"] in connected]
    print(f"  {title}: {len(filtered_nodes)} nodes, {len(filtered_edges)} edges")
    display(build_graph(filtered_nodes, filtered_edges))

print("Filter helpers ready.")
```

## 📊 Step 6 — Degree Centrality Analysis

Identify the most connected (critical) regulatory nodes — these are compliance pinch-points that affect the most other obligations.

```python
import networkx as nx
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

G = nx.DiGraph()
for n in NODES:
    G.add_node(n["id"], **n)
for e in EDGES:
    G.add_edge(e["start"], e["end"], label=e["label"])

degree = dict(G.degree())
sorted_nodes = sorted(degree.items(), key=lambda x: x[1], reverse=True)[:15]
node_lookup = {n["id"]: n for n in NODES}

labels  = [n[0] for n in sorted_nodes]
values  = [n[1] for n in sorted_nodes]
colors  = [CATEGORY_COLORS.get(node_lookup[l]["category"], "#CBD5E1") for l in labels]
readable = [node_lookup[l]["label"].replace("\n", " ") for l in labels]

fig, ax = plt.subplots(figsize=(14, 7))
fig.patch.set_facecolor("#0F172A")
ax.set_facecolor("#1E293B")

bars = ax.barh(range(len(labels)), values, color=colors, edgecolor="#334155", linewidth=0.8)
ax.set_yticks(range(len(labels)))
ax.set_yticklabels(readable, color="#E2E8F0", fontsize=10)
ax.set_xlabel("Degree Centrality (total connections)", color="#94A3B8")
ax.set_title("Top 15 Most Connected Regulatory Nodes", color="#F8FAFC",
             fontsize=14, fontweight="bold", pad=15)
ax.tick_params(colors="#94A3B8")
for spine in ["bottom","left"]:
    ax.spines[spine].set_color("#334155")
for spine in ["top","right"]:
    ax.spines[spine].set_visible(False)
ax.invert_yaxis()

for bar, val in zip(bars, values):
    ax.text(bar.get_width() + 0.1, bar.get_y() + bar.get_height()/2,
            str(val), va="center", color="#94A3B8", fontsize=9)

patches = [mpatches.Patch(color=c, label=k) for k, c in CATEGORY_COLORS.items()]
ax.legend(handles=patches, loc="lower right", facecolor="#0F172A",
          edgecolor="#334155", labelcolor="#E2E8F0", fontsize=8)

plt.tight_layout()
plt.savefig("centrality_analysis.png", dpi=150, bbox_inches="tight", facecolor="#0F172A")
plt.show()
print()
print("KEY INSIGHT: Nodes with highest degree are compliance pinch-points.")
print("  EU_ART16 (Provider Obligations) is the gateway to all other requirements.")
print("  GDPR_A22 (Automated Decisions) is critical for both AI systems.")
```
